```python
model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
avis_utilisateur = {
    "aime": ["Dark", "Breaking Bad", "Black Mirror"],
    "neutre": ["Lupin"],
    "naime_pas": ["Elite", "Riverdale"]
}
```

```python
def construire_prompt(avis):
    return f"""
Tu es un moteur de recommandation de séries TV.

Analyse les préférences suivantes :
- Séries aimées : {', '.join(avis['aime'])}
- Séries neutres : {', '.join(avis['neutre'])}
- Séries non aimées : {', '.join(avis['naime_pas'])}

Objectif :
Proposer 5 séries pertinentes.

Contraintes STRICTES :
- uniquement des séries TV
- éviter les séries déjà mentionnées
- explication courte (max 15 mots)
- réponse UNIQUEMENT en JSON

Format attendu :
[
  {{"nom": "Nom de la série", "explication": "raison courte"}}
]
"""
```

```python
def parse_response(text):
    try:
        return json.loads(text)
    except:
        start = text.find("[")
        end = text.rfind("]") + 1
        return json.loads(text[start:end])
```

```python
def generer_recommandations(model, avis):
    prompt = construire_prompt(avis)
    response = model.generate_content(prompt)
    return parse_response(response.text)
```

```python
avis_test = {
    "aime": ["Dark", "Black Mirror"],
    "neutre": ["Lupin"],
    "naime_pas": ["Elite"]
}

print(generer_recommandations(model, avis_test))
```
